```python
mileage=[]
```

```python
for i in main_links:
    con1=requests.get(f'https://www.bikewale.com/{i}')
    text = con1.text
    html_readable_format = BeautifulSoup(text)
    kmpl=html_readable_format.find_all('td',class_="o-eqqVmt o-eemiLE o-cYdrZi o-eZTujG")[1].text
    if kmpl.endswith('kmpl'):
        mileage.append(kmpl)
    else:
        mileage.append(np.nan)
```

```python
weight=[]
```

```python
for i in main_links:
    con1=requests.get(f'https://www.bikewale.com/{i}')
    text = con1.text
    html_readable_format = BeautifulSoup(text)
    kg1=html_readable_format.find_all('td',class_="o-eqqVmt o-eemiLE o-cYdrZi o-eZTujG")[3].text
    kg2=(html_readable_format.find_all('td',class_="o-eqqVmt o-eemiLE o-cYdrZi o-eZTujG")[2].text)
    if kg1.endswith('kg'):
        weight.append(kg1)
    else:
        weight.append(kg2)
```

```python
weight
```

```python
fuel=[]
```

```python
for i in main_links:
    con1=requests.get(f'https://www.bikewale.com/{i}')
    text = con1.text
    html_readable_format = BeautifulSoup(text)
    try:
        l1=html_readable_format.find_all('td',class_="o-eqqVmt o-eemiLE o-cYdrZi o-eZTujG")[4].text
        l2=html_readable_format.find_all('td',class_="o-eqqVmt o-eemiLE o-cYdrZi o-eZTujG")[3].text
        if l1.endswith('tres'):
            fuel.append(l1)
        else:
             fuel.append(l2)
    except IndexError:
            fuel.append(np.nan)
```

```python
len(fuel)
```

```python
top_speed=[]
```

```python
for i in main_links:
    con1=requests.get(f'https://www.bikewale.com/{i}')
    text = con1.text
    html_readable_format = BeautifulSoup(text)
    s1=html_readable_format.find_all('div',class_="mLrdoz lI0JO_")[5].text
    s2=html_readable_format.find_all('div',class_="mLrdoz lI0JO_")[6].text
    if s1.endswith('h'):
        top_speed.append(s1)
    else:
        top_speed.append(s2)
```
